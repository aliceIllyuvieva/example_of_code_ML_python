```
from IPython.core.interactiveshell import InteractiveShell 
InteractiveShell.ast_node_interactivity = "all"

import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.model_selection import train_test_split as tts 
from sklearn import datasets
from sklearn.model_selection import KFold
import random
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
import itertools
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from collections import OrderedDict
```

```
# 2 number
file = 'http://web.stanford.edu/~oleg2/hse/heart/Heart.csv'
df = pd.read_csv(file)
df = df.dropna()

df['AHD_new'] = (df['AHD'] == 'Yes').astype(int)

thal_to_int = {'fixed': 0, 'normal' : 1, 'reversable' : 2}
for row in ['Thal'] :
  df = df.replace({str(row): thal_to_int})

chestpain_to_int = {'typical': 0, 'asymptomatic' : 1, 'nonanginal' : 2, 'nontypical' : 3}
for row in ['ChestPain'] :
  df = df.replace({str(row): chestpain_to_int})

X = df[['Age','RestBP','Chol','Fbs','RestECG','MaxHR','ExAng','Oldpeak','Slope','Ca', 'Sex', 'Thal', 'ChestPain']]
y = df['AHD_new']

X_train, X_test, y_train, y_test = tts(X, y, test_size = 0.4)

oob_errors_forest = []
test_errors_forest = []

oob_errors_bagging = []
test_errors_bagging = []

test_errors_single_tree = []



estimators = range(5, 300, 2)
model = RandomForestClassifier(n_estimators = estimators, oob_score=True,
                               max_features="sqrt",
                               random_state=0, n_jobs=-1) 
model_2 = DecisionTreeClassifier()
model_3 = DecisionTreeClassifier(max_depth=1)

for n_estimators in estimators:
  model.set_params(n_estimators=n_estimators)
  model.fit(X_train, y_train)
  predict = model.predict(X_test)

  test_error_forest = np.mean((y_test - predict)**2)
  test_errors_forest.append(test_error_forest)
  oob_error_forest = 1 - model.oob_score_
  oob_errors_forest.append(oob_error_forest)

  model_2 = BaggingClassifier(oob_score=True)
  model_2.set_params(n_estimators=n_estimators)
  model_2.fit(X_train, y_train)
  predict = model_2.predict(X_test)

  test_error_bagging = np.mean((y_test - predict)**2)
  test_errors_bagging.append(test_error_bagging)
  oob_error_bagging = 1 - model_2.oob_score_
  oob_errors_bagging.append(oob_error_bagging)

  model_3.fit(X_train, y_train)
  predict = model_3.predict(X_test)
  test_error = np.mean((y_test - predict)**2)
  test_errors_single_tree.append(test_error)



plt.figure(figsize=(7,7))
plt.plot(estimators, test_errors_forest, label = 'Test error : Forest', c='#FFD700')
plt.plot(estimators, oob_errors_forest, label = 'OOB error : Forest', c='blue')
plt.plot(estimators, test_errors_bagging, label = 'Test error : Bagging', c='black')
plt.plot(estimators, oob_errors_bagging, label = 'OOB error : Bagging', c='green')
plt.plot(estimators, test_errors_single_tree, linestyle='dashed', c='black')
plt.xlabel("Number of trees")
plt.ylabel("Error")
plt.legend(loc="upper right")
plt.ylim(0.1, 0.3, 0.05)
plt.show()


```

```
# 4 number
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.metrics import accuracy_score
sURL = 'http://web.stanford.edu/~oleg2/hse/14-cancer-microarray/14cancer.'
X_train = pd.read_csv(sURL+'xtrain.zip', delimiter='\s+', header=None).T
X_test = pd.read_csv(sURL+'xtest.zip', delimiter='\s+', header=None).T
y_train = pd.read_csv(sURL+'ytrain', delimiter='\s+', header=None).values[0]
y_test = pd.read_csv(sURL+'ytest', delimiter='\s+', header=None).values[0]

X_train = X_train.iloc[:, :500]
X_test = X_test.iloc[:, :500]


estimators = range(1, 500, 5)
model1 = RandomForestClassifier(n_estimators = estimators, random_state=0, n_jobs=-1, max_features=500)
model2 = RandomForestClassifier(n_estimators = estimators, random_state=1, n_jobs=-1, max_features=250)
model3 = RandomForestClassifier(n_estimators = estimators, random_state=2, n_jobs=-1, max_features='sqrt')

test_errors1 = []
test_errors2 = []
test_errors3 = []


for n_estimators in estimators:
  model1.set_params(n_estimators=n_estimators)
  model1.fit(X_train, y_train)
  predict = model1.predict(X_test)

  test_error1 = np.mean((y_test - predict)**2)
  test_errors1.append(test_error1/100)

  model2.set_params(n_estimators=n_estimators)
  model2.fit(X_train, y_train)
  predict = model2.predict(X_test)

  test_error2 = np.mean((y_test - predict)**2)
  test_errors2.append(test_error2/100)

  model3.set_params(n_estimators=n_estimators)
  model3.fit(X_train, y_train)
  predict = model3.predict(X_test)

  test_error3 = np.mean((y_test - predict)**2)
  test_errors3.append(test_error3/100)

print(test_errors1)
print(test_errors2)
print(test_errors3) 

plt.figure(figsize=(8,7))
plt.plot(estimators, test_errors1, label = 'm = p', c='yellow')
plt.plot(estimators, test_errors2, label = 'm = p/2', c='blue')
plt.plot(estimators, test_errors3, label = 'm = srt(p)', c='green')
plt.xlabel("Number of trees")
plt.ylabel("Test Classification Error")
plt.legend(loc="upper right")
plt.ylim(0.1, 0.7)
plt.show()
```
